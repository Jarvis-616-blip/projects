### Objective: Practice data visualization techniques for better data representation.

```python
import matplotlib.pyplot as plt
# 1. Line Plot
x = [1,2,3,4,5]
y = [10,15,25,30,50]
plt.plot(x, y, marker='o')
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
# 2. Bar Graph
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75,85,60,90]
plt.bar(students, marks, color=['blue','green','orange','red'])
plt.title('Marks of Students')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()
```

```python
# 3. Pie Chart
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45,25,20,10]
explode = [0.1, 0, 0, 0]  # highlight highest revenue
plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%')
plt.title('Revenue Distribution by Region')
plt.show()
```
