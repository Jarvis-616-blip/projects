# Assignment 1 - Working with NumPy
### Objective: Understand the basics of NumPy, array creation, and manipulation.

```python
import numpy as np
# 1D Array from 1 to 20
arr = np.arange(1, 21)
print('Array:', arr)
print('Sum:', np.sum(arr))
print('Mean:', np.mean(arr))
print('Median:', np.median(arr))
print('Standard Deviation:', np.std(arr))
print('Indices > 10:', np.where(arr > 10))
```

```python
# 2D Array 4x4
arr2d = np.arange(1, 17).reshape(4,4)
print('Array:
', arr2d)
print('Transpose:
', arr2d.T)
print('Row-wise sum:', np.sum(arr2d, axis=1))
print('Column-wise sum:', np.sum(arr2d, axis=0))
```

```python
# 3x3 Random Arrays
a = np.random.randint(1, 21, (3,3))
b = np.random.randint(1, 21, (3,3))
print('A:
', a)
print('B:
', b)
print('Addition:
', a+b)
print('Subtraction:
', a-b)
print('Multiplication:
', a*b)
print('Dot Product:
', np.dot(a,b))
```

```python
# Reshape and Slice
arr3 = np.arange(1, 13).reshape(3,4)
print('Original Array:
', arr3)
print('Sliced (first two rows, last two columns):
', arr3[:2, -2:])
```
