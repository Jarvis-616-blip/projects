```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import datasets
df = pd.read_csv("covid19_Confirmed_dataset.csv")
df.head()
```

```python
df.shape
```

```python
#delete useless columns 
df1 = df.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
df.head()
```

```python
#aggregate the row by the country
corona = df.groupby("Country/Region").sum()
corona.head()
```

```python
corona.shape
```

```python
#Visulization
corona.loc["Afghanistan"].plot()
corona.loc["India"].plot()
corona.loc["Spain"].plot()
plt.legend()
```

```python
#calculate a good measure
corona.loc["Spain"].plot()
```

```python
corona.loc["India"][:3].plot()
```

```python
#calcualte the firt deriavtive 
corona.loc["India"].diff().plot()
```

```python
#Maximum infection rate
corona.loc["India"].diff().max()
```

```python
country = list(corona.index)
max_infect = []


for c in country:
    val = corona.loc[c].apply(pd.to_numeric, errors='coerce').diff().max()
    max_infect.append(0 if pd.isna(val) else val)
corona["Max_infecrate"]= max_infect


corona
```

```python
corona1 = pd.DataFrame(corona["Max_infecrate"])
```

```python
corona1
```

```python
#importing dataset 2
happy = pd.read_csv("worldwide_happiness_report.csv")
```

```python
happy.head()
```

```python
#dropiing useless column 
useless = ["Overall rank","Score","Generosity","Perceptions of corruption"]
```

```python
happy.drop(useless,axis=1,inplace=True)
```

```python
happy.head()
```

```python
happy.set_index("Country or region",inplace=True)
```

```python
happy.head()
```

```python
#join the datsets
corona1.shape
```

```python
happy.shape
```

```python
df2 = corona1.join(happy,how="inner")
df2
```

```python
df2.corr()
```

```python
df2
```

```python
#visulaization 
df2["log_gdp"] = np.log(df2["GDP per capita"])
df2["log_infrate"] = np.log(df2["Max_infecrate"])

# Plot regplot on transformed columns
sns.scatterplot(
    data=df2,
    x="log_gdp",
    y="log_infrate"
)

plt.xlabel("Log GDP per Capita")
plt.ylabel("Log Max Infection Rate")
plt.title("Log-Log Regression: GDP vs Infection Rate")

plt.show()
```

```python
df_clean = df2[(df2["GDP per capita"] > 0) & (df2["Max_infecrate"] > 0)].copy()

# Log transform
df_clean["log_gdp"] = np.log(df_clean["GDP per capita"])
df_clean["log_inf"] = np.log(df_clean["Max_infecrate"])

# REGPLOT with line
sns.regplot(data=df_clean, x="log_gdp", y="log_inf")

plt.xlabel("Log GDP per Capita")
plt.ylabel("Log Max Infection Rate")
plt.title("Log-Log Regression")
plt.show()
```

```python
df2["log_inf"] = np.log(df2["Max_infecrate"])

# Scatter plot
sns.scatterplot(
    data=df2,
    x="Social support",
    y="log_inf"
)

plt.xlabel("Social Support")
plt.ylabel("Log(Max Infection Rate)")
plt.title("Social Support vs Log Infection Rate")
plt.show()
```

```python
df_clean = df2[df2["Max_infecrate"] > 0].copy()

# Log transform Y
df_clean["log_inf"] = np.log(df_clean["Max_infecrate"])

# REGPLOT
sns.regplot(
    data=df_clean,
    x="Social support",
    y="log_inf"
)

plt.xlabel("Social Support")
plt.ylabel("Log(Max Infection Rate)")
plt.title("Social Support vs Log Infection Rate (Regression)")
plt.show()
```
