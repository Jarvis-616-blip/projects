```python
df = pd.read_csv("austin_weather.csv")
df
```

```python
#drop unecessory columns
df = df.drop(["Events","Date","SeaLevelPressureLowInches"],axis=1)
```

```python
df =df.replace('T',0.0)
df =df.replace('-',0.0)
```

```python
x = df.drop(['PrecipitationSumInches'],axis=1)
```

```python
y = df["PrecipitationSumInches"]
```

```python
#reshaping into 2d vector
y = y.values.reshape(-1,1)
```

```python
day_index = 798
days = [i for i in range(y.size)]
```

```python
#intialize linear regression classifier
clf = LinearRegression()

clf.fit(x,y)
```

```python

#plot graph
print("The Precipitaion treand graph: ")
# flatten y to a 1D float array so matplotlib receives scalar y-values (not 1-element arrays)
y_flat = np.asarray(y, dtype=float).ravel()

plt.scatter(days, y_flat, color='g', label='Daily Precipitation')
plt.scatter(days[day_index], y_flat[day_index], color='r', label='Selected Day')
plt.title("precipitation Level")
plt.xlabel("Days")
plt.ylabel("Precipitaion in inches")
plt.legend()
plt.show()
X_vis = x.filter(['TempAvgF','DewPointAvgF','HumidityAvgPercent','SeaLevelPressureAvgInches','VisibilityAvgMiles','WindAvgMPH'])
```

```python
print("The Precipitation trend graph: ")

for i, col in enumerate(X_vis.columns):
    plt.subplot(3, 2, i + 1)
    # Convert column values to numeric (coerce non-numeric to NaN)
    y_values = pd.to_numeric(X_vis[col], errors='coerce')
    # Plot the daily values and highlight the selected day
    plt.scatter(days, y_values, color='g')
    plt.scatter(days[day_index], y_values.iloc[day_index], color='r')
    # Ensure the title is a string (avoids passing a float)
    plt.title(str(col))

plt.tight_layout()
plt.show()
    
```
